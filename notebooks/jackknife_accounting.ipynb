```python
%matplotlib inline
import matplotlib.pyplot as plt
import treecorr
import numpy as np
```

```python
N = 1000
cat = np.array([np.random.rand(N),np.random.rand(N)]).T
```

```python
catA = cat[cat[:,0] < .5]
catB = cat[cat[:,0] > .5]
```

```python
plt.scatter(cat[:,0],cat[:,1])
plt.figure()
plt.scatter(catA[:,0],catA[:,1],color='g')
plt.scatter(catB[:,0],catB[:,1],color='r')
```

```python
nnn = treecorr.NNNCorrelation(min_sep=.2,max_sep=.3,nbins=1,min_u=0,max_u=1,min_v=-1,max_v=1,nubins=1,nvbins=1,bin_slop=0)
cat = treecorr.Catalog(x=cat[:,0],y=cat[:,1])
nnn.process(cat,cat,cat)
```

OK, here's our target number:

```python
target = nnn.ntri
print target
```

## Counting each region separately

```python
catA = treecorr.Catalog(x=catA[:,0],y=catA[:,1])
catB = treecorr.Catalog(x=catB[:,0],y=catB[:,1])
```

```python
nnnA = treecorr.NNNCorrelation(min_sep=.2,max_sep=.3,nbins=1,min_u=0,max_u=1,min_v=-1,max_v=1,nubins=1,nvbins=1,bin_slop=.1)
nnnB = treecorr.NNNCorrelation(min_sep=.2,max_sep=.3,nbins=1,min_u=0,max_u=1,min_v=-1,max_v=1,nubins=1,nvbins=1,bin_slop=.1)
nnnA.process(catA,catA,catA)
nnnB.process(catB,catB,catB)
```

```python
print nnnA.ntri
print nnnB.ntri
print (nnnA.ntri + nnnB.ntri)
```
