```python
%matplotlib inline
import matplotlib.pyplot as plt
from astropy.io import fits
import numpy as np
```

```python
from glob import glob
```

```python
halofiles = glob('/u/ki/jderose/public_html/bcc/catalog/halos/semihemisphere/buzzard/flock/buzzard-0/*.fits')
```

```python
ralist = []
declist = []
zlist = []
for halofile in halofiles:
    halos = fits.getdata(halofile)
    ralist.append(halos['RA'])
    declist.append(halos['DEC'])
    zlist.append(halos['Z'])
```

```python
halos = fits.getdata('/u/ki/jderose/public_html/bcc/catalog/halos/semihemisphere/buzzard/flock/buzzard-0/Chinchilla-0_halos.0.fits')
halos2 = fits.getdata('/u/ki/jderose/public_html/bcc/catalog/halos/semihemisphere/buzzard/flock/buzzard-0/Chinchilla-0_halos.1.fits')
```

```python
print len(halos)
print len(halos2['RA'])
print halos.columns
```

```python
merged_ras = np.concatenate(ralist)
merged_decs = np.concatenate(declist)
merged_zs = np.concatenate(zlist)
```

```python
counts2,_,_,_ = plt.hist2d(merged_ras,merged_decs,bins=200)
plt.figure()
plt.hist(merged_zs,bins=50);
```

```python
counts2,_,_,_ = plt.hist2d(halos.RA,halos.DEC,bins=200)
plt.figure()
plt.hist(halos.Z,bins=50);
```

```python
counts2,_,_,_ = plt.hist2d(halos2.RA,halos2.DEC,bins=200)
plt.figure()
plt.hist(halos2.Z,bins=50);
```
